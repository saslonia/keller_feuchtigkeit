### Über die Interpretation der Messwerte

#### Feuchtigkeit der Wände

Laut [dieser Quelle](https://www.ausbaupraxis.de/feucht-oder-froehlich-die-feuchtemessung-17112021) sind die Messwerte wie folgt zu bewerten. In den folgenden Diagrammen werden die Kategorien auch farblich entsprechend gekennzeichnet.

- bis 40 Skalenanteile: trockener Baustoff --> grün
- 40 – 80 Skalenanteile: mäßig feucht (Material-Ausgleichsfeuchte) --> gelb
- 80 – 100 Skalenanteile: erhöhte Baustofffeuchte --> orange
- 100 – 150 Skalenanteile: durchfeuchtetes Bauteil --> rot (ab 150 Richtung schwarz)

> ❗**Achtung:** Die Messwerte können durch andere Materialien in der unmittelbaren Nähe der Messung stark beeinflusst werden.
> Diese Materialien erzeugen einen höheren Messwert:
> - Metall
> - Salz
> - Beton (Boden und Decke des Raumes)

#### Diagramme

- Alle Diagramme haben dieselbe Farbskala, damit sie miteinander vergleichbar sind.
- In den Diagrammen sind kleine graue Kreuze zu finden. Diese markieren die Orte der tatsächlichen Messungen.
- Die Bereiche zwischen den Messpunkten sind linear interpoliert (es wird ein linearer Verlauf zwischen den Messpunkten angenommen)

## Handlungsempfehlung

1. Aussalzungen behandeln: Oberflächliches Salz abkrazen, zusätzlich Opferputz auftragen. Eventuell auch Operputz bei dem Hausabfluss auftragen.
2. Wasser mit Bautrocknern aus dem Keller entfernen, durchgängiger Betrieb von ca. 2 Wochen (tatsächliche Betriebszeit von Luftmesswerten abhängig machen)
    - Mindestens 2 Bautrockner, die jeweils mindestens 50 Liter, besser 100 Liter innerhalb von 24 Stunden aus der Luft ziehen können. Einer im Bereich von Keller 8 und 9, einer im Bereich von Keller 1 und 5. 
    - Wasser direkt in den Abfluss abführen, nicht im internen Behälter des Trockners sammeln
    - Dabei alle Türen der einzelnen Kellerräume öffnen und in den Räumen alle Gegenstände mindestens 10 cm von den Wänden abrücken, von den Außenwänden 20 cm
    - Zusätzlich Ventilatoren aufstellen, die für Luftbewegung sorgen, damit die Feuchtigkeit auch aus der allerletzten Ecke herauskommt.
    - Zusätzlich einen Kondenstrockner (ca. 20 Liter innerhalb von 24 Stunden) in der kritischen Ecke des Fahrradkellers betreiben
3. Feuchtigkeitsmessung der kritischen Wandstellen (für spätere Erfolgskontrolle)
4. Zwei bis vier Wochen warten, damit das Wasser aus den äußeren Wandschichten in die inneren Wandschichten nachströmen kann
5. Feuchtigkeitsmessung der kritischen Wandstellen (für spätere Erfolgskontrolle)
6. Erneuter Betrieb der Bautrockner (wie Punkt 2)
7. Feuchtigkeitsmessung der kritischen Wandstellen (für spätere Erfolgskontrolle)
8. Ca. 1 Monat warten
9. Feuchtigkeitsmessung der kritischen Wandstellen, wenn diese wieder feuchter sind, haben wir mit einer sehr hohen Wahrscheinlichkeit eine undichte Stelle in der Kellerwand.
